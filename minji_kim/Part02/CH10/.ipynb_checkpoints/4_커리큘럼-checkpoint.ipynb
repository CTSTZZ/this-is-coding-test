```python
n = int(input())
data = []
for _ in range(n) :
    data.append(list(map(int, input().split()))) # , 
```

```python
from collections import deque
import copy

#     0 
indegree = [0] * (n+1)
#          
graph = [[0] for i in range(n+1)]
#    0 
time = [0] * (n+1)

for i in range(1, n+1) :
    time[i] = data[i-1][0] #       
    
    for j in data[i-1][1:-1] : # 
        indegree[i] += 1  #   [0, 0, 1, 1, 2, 1]
        
        graph[j].append(i) #  j() i 
```

```python
graph # 1 2, 3, 4  / 3  4, 5 
```

```python
#   
def topology_sort() :
    result = copy.deepcopy(time) # time  
    q = deque()
    
    for i in range(1, n+1) :
        if indegree[i] == 0 :
            q.append(i)  # deque([1])
    
    # q   
    while q :
        now = q.popleft()  #   ; 1
        
        #      1
        for i in graph[now] :
            result[i] = max(result[i], result[now] + time[i])
            indegree[i] -= 1
            
            #   0    
            if indegree[i] == 0 :
                q.append(i)
                
    for i in range(1, n+1) :
        print(result[i])
```

```python
topology_sort()
```
